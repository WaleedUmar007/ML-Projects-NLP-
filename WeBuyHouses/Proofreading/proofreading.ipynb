```python
#Code to proofreead the text

def proofread_text(text):
    model_engine = 'davinci'
    prompt = (f'Please proofread the following text:\n{text}\n\n'
              f'The corrected text is:')
    response = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )
    return response.choices[0].text.strip()
```
